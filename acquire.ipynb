```python
def get_events(cursor='', event_type='successful', **kwargs):
    url = "https://api.opensea.io/api/v1/events"
    query = {"only_opensea": "false", 
             "event_type": event_type,
             "cursor": cursor,
             **kwargs
             }

    headers = {
        "Accept": "application/json",
        "X-API-KEY": opensea_apikey  # Use the variable you defined
    }
    response = requests.request("GET", url, headers=headers, params=query)

    return response.json()
```

```python
data = get_events(event_type='successful', occurred_before=end_date, occurred_after=start_date, cursor='')
data
```
